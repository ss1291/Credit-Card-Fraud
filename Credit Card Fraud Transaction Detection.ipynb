```python
from imblearn.over_sampling import RandomOverSampler

# Assuming X is your feature matrix and Y is your target vector
print("Type of X:", type(X))
print("Type of Y:", type(Y))

# Check if X and Y are not None
if X is not None and Y is not None:
    # Check if X and Y are non-empty
    if not X.empty and not Y.empty:
        # Convert Y to numpy array
        Y_array = np.array(Y)

        # Print some information about the data
        print("Shape of X:", X.shape)
        print("Number of classes in Y:", np.unique(Y_array))

        # Create an instance of RandomOverSampler
        ros = RandomOverSampler(random_state=42)

        try:
            # Apply RandomOverSampler to generate synthetic samples
            x_resampled, y_resampled = ros.fit_resample(X, Y_array)
            print("RandomOverSampler applied successfully.")
            # Check the shape of resampled data
            print("Shape of resampled data:")
            print("x_resampled:", x_resampled.shape)
            print("y_resampled:", y_resampled.shape)
        except Exception as e:
            print("Error occurred while applying RandomOverSampler:", e)
            print("Detailed Error:", str(e))
    else:
        print("Error: X or Y is empty. Cannot proceed with sampling.")
else:
    print("Error: X or Y is None. Cannot proceed with sampling.")
```

# Model Building

The models will be based on Logistic Regression

```python
#Build the Logistic Regression model:
from sklearn.linear_model import LogisticRegression
logistic_model = LogisticRegression(random_state=42)
logistic_model.fit(X_train, Y_train)
```
